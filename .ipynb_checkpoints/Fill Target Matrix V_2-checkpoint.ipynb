```python
class Course:
    def __init__(self, id, name, no_of_hours, no_of_valid_rooms, valid_rooms, max_no_hours_per_day, max_consecutive_hours, no_of_hours_to_schedule_when_assigning, class_flag, batch_flag):
        self.id = id
        self.name = name
        self.no_of_hours = no_of_hours
        self.no_of_valid_rooms= no_of_valid_rooms
        self.valid_rooms = valid_rooms
        self.max_no_hours_per_day = max_no_hours_per_day
        self.max_consecutive_hours = max_consecutive_hours
        self.no_of_hours_to_schedule_when_assigning = no_of_hours_to_schedule_when_assigning
        self.class_flag = class_flag
        self.batch_flag = batch_flag
        
class Student_Group:
    def __init__(self, id, name, courses):
        self.id = id
        self.name = name
        self.courses = courses

class Lecturer:
    def __init__(self, id, name, department, max_no_hours_per_day, max_no_hours_per_week, max_consecutive_hours, rank, availability, courses):
        self.id = id
        self.name = name
        self.department = department
        self.max_no_hours_per_day = max_no_hours_per_day
        self.max_no_hours_per_week = max_no_hours_per_week
        self.max_consecutive_hours = max_consecutive_hours
        self.rank = rank
        self.availability = availability
        self.courses = courses
        
class Room:
    def __init__(self, id, name):
        self.id = id
        self.name = name
class Day:
    def __init__(self, id, name):
        self.id = id
        self.name = name
class Hour:
    def __init__(self, id, name):
        self.id = id
        self.name = name
```

```python
course_list = []
lecturer_list = []
student_group_list = []
rooms_list = []
days_list = []
hours_list = []
no_of_days = -1
no_of_hours = -1

with open("input.txt", "r") as f:
    lines = f.readlines()
    i = 0
    while(i<len(lines)):
        lines[i] = lines[i][:-1]
        print("i=",i)
        if lines[i] == "Days":
            i+=1
            no_of_days = int(lines[i][:-1])
            i+=1
            id = 0
            for j in range(i, i+no_of_days):
                name = lines[j][:-1]
                days_list.append(Day(id, name))
                id+=1
            i+=no_of_days
        elif lines[i] == "Hours":
            i+=1
            no_of_hours = int(lines[i][:-1])
            i+=1
            id = 0
            for j in range(i, i+no_of_hours):
                name = lines[j][:-1]
                hours_list.append(Hour(id, name))
                id+=1
            i+=no_of_hours
        elif lines[i] == "Courses":
            i+=1
            number_of_courses = int(lines[i][:-1])
            i+=1
            count = 0
            while(count<number_of_courses):
                course_id, course_name = lines[i][:-1].split(" ", 1)
                print(course_id, course_name)
                i+=1
                class_flag = bool(lines[i][:-1]=="True")
                i+=1
                batch_flag = bool(lines[i][:-1]=="True")
                i+=1
                no_of_hours_to_schedule_when_assigning = int(lines[i][:-1])
                i+=1
                max_no_hours_per_day = int(lines[i][:-1])
                i+=1
                no_hours_per_week = int(lines[i][:-1])
                i+=1
                max_consecutive_hours = int(lines[i][:-1])
                i+=1
                no_of_valid_rooms = int(lines[i][:-1])
                i+=1
                valid_rooms = []
                for j in range(i, i+no_of_valid_rooms):
                    valid_rooms.append(lines[j][:-1])
                i+=no_of_valid_rooms
                course_list.append(Course(course_id, course_name, no_hours_per_week, no_of_valid_rooms, valid_rooms, max_no_hours_per_day, max_consecutive_hours, no_of_hours_to_schedule_when_assigning, class_flag, batch_flag))
                count+=1
        elif lines[i] == "Rooms":
            i+=1
            number_of_rooms = int(lines[i][:-1])
            i+=1
            id = 0
            for j in range(i, i+number_of_rooms):
                name = lines[j][:-1]
                rooms_list.append(Room(id, name))
                id+=1
            i+=number_of_rooms
        elif lines[i] == "Student Groups":
            i+=1
            number_of_student_groups = int(lines[i][:-1])
            i+=1
            id = 0
            while(id<number_of_student_groups):
                name = lines[i][:-1]
                i+=1
                number_of_courses = int(lines[i][:-1])
                i+=1
                course_ids = []
                while(number_of_courses>0):
                    print(lines[i][:-1].split())
                    course = lines[i][:-1]
                    course_ids.append(course)
                    i+=1
                    number_of_courses-=1
                student_group_list.append(Student_Group(id, name, course_ids))
                id+=1
            
        elif lines[i] == "Lecturers":
            i+=1
            number_of_lecturers = int(lines[i][:-1])
            i+=1
            id = 0
            while(id<number_of_lecturers):
                name = lines[i][:-1]
                i+=1
                department = lines[i][:-1]
                i+=1
                max_no_hours_per_day = int(lines[i][:-1])
                i+=1
                max_no_hours_per_week = int(lines[i][:-1])
                i+=1
                max_consecutive_hours = int(lines[i][:-1])
                i+=1
                rank = lines[i][:-1]
                i+=1
                availability = [int(x) for x in lines[i][:-1].split()]
                i+=1
                no_of_courses = int(lines[i][:-1])
                i+=1
                course_ids = []
                while(no_of_courses>0):
                    print(lines[i][:-1])
                    course = lines[i][:-1]
                    course_ids.append(course)
                    i+=1
                    no_of_courses-=1
                lecturer_list.append(Lecturer(id, name, department, max_no_hours_per_day, max_no_hours_per_week, max_consecutive_hours, rank, availability, course_ids))
                id+=1
        else:
            i+=1
```

```python
def no_batch_conflicts(target_matrix, rows, columns, day, hour, batch, student_group_id, course_id, lecturer_id):
    for i in range(len(rows)):
        current_day = rows[i]["day"]
        current_hour = rows[i]["hour"]
        if current_day == day and current_hour == hour:
            for j in range(len(columns)):
                if columns[j]["subset"] == "batch" and columns[j]["student_group"] == student_group_id and target_matrix[i][j]==1:
                    if columns[j]["batch"] == batch or columns[j]["lecturer"] == lecturer_id or columns[j]["course"]==course_id: 
                        return False
    return True
```

```python
import numpy as np
def fillTargetMatrix(column_numbers, columns, hours_for_columns, rows, day_dictionary, hour_dictionary, room_dictionary, lecturer_dictionary, student_group_dictionary, course_dictionary, no_of_days, no_of_hours):
    target_matrix = np.zeros((len(rows), len(columns)))
    for column_number in column_numbers:
        print(column_number)
        if columns[column_number]["subset"] == "all":
#             if input("all") == "N":
#                 return target_matrix
            course_id = columns[column_number]['course']
            lecturer_id = columns[column_number]['lecturer']
            student_group_id = columns[column_number]['student_group']
            no_of_hours_to_schedule = hours_for_columns[column_number]
            no_of_hours_to_schedule_when_assigning = course_dictionary[course_id].no_of_hours_to_schedule_when_assigning
            while no_of_hours_to_schedule > 0:
                valid_rows = []
                day_flag = [False] * no_of_days
                for day in range(no_of_days):
                    course_count = 0
                    lecturer_count = 0
                    for i in range(len(rows)):
                        for j in range(len(columns)):
                            if target_matrix[i][j] == 1:
                                if rows[i]["day"] == day and columns[j]["lecturer"] == lecturer_id:
                                    lecturer_count+=1
                                if rows[i]["day"] == day and columns[j]["course"] == course_id:
                                    course_count+=1
                    if lecturer_dictionary[lecturer_id].max_no_hours_per_day > lecturer_count or course_dictionary[course_id].max_no_hours_per_day > course_count:
                        day_flag[day] = True

                # For Valid rows following conditions:
                # 1. Check if slot is free, i.e. 0
                # 2. Check room validity using atribute valid_rooms in 'Course' object
                # 3. Check Hour and Day validity using attribute availability in 'Lecturer' object
                for i in range(len(rows)):
                    day = rows[i]["day"]
                    hour = rows[i]["hour"]
                    if target_matrix[i][column_number] == 0 and day_flag[day]:
                        if room_dictionary[rows[i]["room"]].name in course_dictionary[course_id].valid_rooms:
                            flag = True
                            for j in range(no_of_hours_to_schedule_when_assigning):
                                if rows[i]['hour'] + j != rows[i+j]['hour'] or lecturer_dictionary[lecturer_id].availability[day*no_of_hours+hour+j] != 1 or target_matrix[i+j][column_number]!=0:
#                                 if target_matrix[i+j][column_number]!=0:
                                    flag = False
                                    break
                            if flag:
                                valid_rows.append(i)


                # 4. Check max no of hours per day of Course and Lecturer
                # 5. Check max consecutive hours of Course and Lecturer
                if len(valid_rows) == 0:
                    print("COULD NOT SCHEDULE")
                    print(column_number, no_of_hours_to_schedule)
                    break
                print(valid_rows)
#                 if input() == "N": 
#                     return target_matrix

                # From valid rows, schedule lecture on first row id

                # Bottom Line randomly chooses row to schedule
                row_id_to_schedule = random.choice(valid_rows)
                # Bottom Line chooses first element
#                 row_id_to_schedule = valid_rows[0]
                for k in range(no_of_hours_to_schedule_when_assigning):
                    day = rows[row_id_to_schedule+k]["day"]
                    hour = rows[row_id_to_schedule+k]["hour"]
                    room = rows[row_id_to_schedule+k]["room"]
                    print(day, hour,room_dictionary[room].name)
#                     if input() == "N":
#                         return target_matrix

                    for j in range(len(columns)):
                        target_matrix[row_id_to_schedule+k][j] = -1

                    for i in range(len(rows)):
                        for j in range(len(columns)):
                            if rows[i]["day"] == day and rows[i]["hour"]==hour and columns[j]["lecturer"] == lecturer_id and target_matrix[i][j]!=1:
                                target_matrix[i][j] = -1
                            if rows[i]["day"] == day and rows[i]["hour"]==hour and columns[j]["student_group"] == student_group_id and target_matrix[i][j]!=1:
                                target_matrix[i][j] = -1   
                    target_matrix[row_id_to_schedule+k][column_number] = 1
                    no_of_hours_to_schedule-=1
        else:
            #Check if another lab has been scheduled, and schedule it at that day and hour if possible.
#             if input("batch") == "N":
#                 return target_matrix
            course_id = columns[column_number]['course']
            lecturer_id = columns[column_number]['lecturer']
            student_group_id = columns[column_number]['student_group']
            no_of_hours_to_schedule = hours_for_columns[column_number]
            batch = columns[column_number]["batch"]
            no_of_hours_to_schedule_when_assigning = course_dictionary[course_id].no_of_hours_to_schedule_when_assigning
            while no_of_hours_to_schedule>0:
                #Find all day and hour available
                valid_rows = []
                for i in range(len(rows)):
                    day = rows[i]["day"]
                    hour = rows[i]["hour"]
                    if room_dictionary[rows[i]["room"]].name in course_dictionary[course_id].valid_rooms:
                        flag = True
                        for j in range(no_of_hours_to_schedule_when_assigning):
                            if rows[i]['hour'] + j != rows[i+j]['hour'] or lecturer_dictionary[lecturer_id].availability[day*no_of_hours+hour+j] != 1 or target_matrix[i+j][column_number]!=0:
#                                 if target_matrix[i+j][column_number]!=0:
                                flag = False
                                break
                        if flag:
                            valid_rows.append(i)
                if len(valid_rows) == 0:
                    print("COULD NOT SCHEDULE")
                    print(column_number, no_of_hours_to_schedule)
                    break
                print(valid_rows)
#                 if input() == "N": 
#                     return target_matrix
                #Valid Rows contains all index that are possible to be scheduled
                scheduled_flag = False
                for i in range(len(rows)):
                    day = rows[i]["day"]
                    hour = rows[i]["hour"]
                    for j in range(len(columns)):
                        #Not checking all batches
                        if columns[j]["subset"] == "batch" and columns[j]["student_group"] == student_group_id:
                            if columns[j]["batch"]!=batch and columns[j]["course"]!=course_id and target_matrix[i][j]==1:
                                batch_has_not_been_scheduled_flag = True
                                if no_batch_conflicts(target_matrix, rows, columns, day, hour, batch, student_group_id, course_id, lecturer_id):
                                    for o in valid_rows:
                                        if rows[o]["day"] == day and rows[o]["hour"] == hour and target_matrix[o][column_number]==0:
                                            print(o)
                                            print("#Schedule Here")
                                            for k in range(no_of_hours_to_schedule_when_assigning):
                                                print(rows[o+k]["day"], rows[o+k]["hour"], room_dictionary[rows[o+k]["room"]].name)
#                                                 if input() == "N":
#                                                     return target_matrix
                                                for l in range(len(columns)):
                                                    target_matrix[o+k][l] = -1
                                                #Make unavailable room and lecturer for same day and time
                                                for m in range(len(rows)):
                                                    for n in range(len(columns)):
                                                        if rows[m]["day"] == rows[o+k]["day"] and rows[m]["hour"]==rows[o+k]["hour"] and columns[n]["lecturer"] == lecturer_id and target_matrix[m][n]!=1:
                                                            target_matrix[m][n] = -1
                                                        if rows[m]["day"] == rows[o+k]["day"] and rows[m]["hour"]==rows[o+k]["hour"] and columns[n]["student_group"] == student_group_id and columns[n]["subset"] == "all" and target_matrix[m][n]!=1:
                                                            target_matrix[m][n] = -1
                                                        if rows[m]["day"] == rows[o+k]["day"] and rows[m]["hour"]==rows[o+k]["hour"] and columns[n]["student_group"] == student_group_id and columns[n]["subset"] == "batch" and columns[n]["batch"] == batch and target_matrix[m][n]!=1:
                                                            target_matrix[m][n] = -1
                                                            
                                                target_matrix[o+k][column_number] = 1
                                                scheduled_flag = True
                                                no_of_hours_to_schedule-=1
                                            break
                                if scheduled_flag:
                                    break
                            if scheduled_flag:
                                break
                            
                #Else Randomly Schedule        
                if scheduled_flag == False:
                    print("Randomly Schedule ")
                    #Schedule new
                    row_id_to_schedule = random.choice(valid_rows)
                    # Bottom Line chooses first element
    #                 row_id_to_schedule = valid_rows[0]
                    scheduled_rows = []
                    for k in range(no_of_hours_to_schedule_when_assigning):
                        day = rows[row_id_to_schedule+k]["day"]
                        hour = rows[row_id_to_schedule+k]["hour"]
                        print(day, hour, room_dictionary[rows[row_id_to_schedule+k]["room"]].name)
#                         if input() == "N":
#                             return target_matrix

                        for j in range(len(columns)):
                            target_matrix[row_id_to_schedule+k][j] = -1

                        for i in range(len(rows)):
                            for j in range(len(columns)):
                                if i not in scheduled_rows:
                                    if rows[i]["day"] == day and rows[i]["hour"]==hour and columns[j]["lecturer"] == lecturer_id and target_matrix[i][j]!=1:
                                        target_matrix[i][j] = -1
                                    if rows[i]["day"] == day and rows[i]["hour"]==hour and columns[j]["student_group"] == student_group_id and columns[j]["subset"] == "all" and target_matrix[i][j]!=1:
                                        target_matrix[i][j] = -1
                                    if rows[i]["day"] == day and rows[i]["hour"]==hour and columns[j]["student_group"] == student_group_id and columns[j]["subset"] == "batch" and columns[j]["batch"] == batch and target_matrix[i][j]!=1:
                                        target_matrix[i][j] = -1
                        
                        target_matrix[row_id_to_schedule+k][column_number] = 1
                        scheduled_rows.append(row_id_to_schedule+k)
                        no_of_hours_to_schedule-=1
                
    return target_matrix
```
